### ***Pyplot API & Object Oriented API***
##### Pyplot API -> Quickly 
##### OO API -> Advance 

```python
x = np.linspace(0,10,1000)
plt.plot(x, np.sin(x), color = "red");
plt.xlabel("Bien x")
plt.ylabel("Sin(x)")
plt.xlim(0,4)
```

```python
x = np.linspace(0,10,1000)
plt.plot(x, np.sin(x), color = "red", linestyle = "dashed");
plt.plot(x, np.cos(x), color = "blue", linestyle = "dashed");
plt.xlabel("Bien x")
plt.ylabel("Sin(x)")
plt.xlim(0,4)
```

```python
x = np.linspace(0,10,1000)
plt.plot(x, np.sin(x), color = "red", linestyle = "dashed", label = "Sin(x)");
plt.plot(x, np.cos(x), color = "blue", linestyle = "dashed", label = "Cos(x)");

plt.title("Sin(X) & Cos(X)")
plt.xlabel("F(x)")
plt.ylabel("Sin(x)")
# plt.xlim(0,4)

plt.legend();
```

```python
x = np.linspace(0,10,1000)
plt.plot(x, np.sin(x), color = "red", linestyle = "dashed", label = "Sin(x)");
plt.plot(x, np.cos(x), color = "blue", linestyle = "dashed", label = "Cos(x)");

plt.title("Sin(X) & Cos(X)")
plt.xlabel("F(x)")
plt.ylabel("Sin(x)")
# plt.xlim(0,4)
# plt.axis([xmin,xmax,ymin,ymax])
plt.axis([0,4,-7,5])

plt.legend();
```

```python
# 00 API
# Prepare new Data
rng = np.random.RandomState(0)

x = rng.randn(100)
y = rng.randn(100)
colors = rng.randn(100)
sizes = 1000*rng.rand(100)

fig, ax = plt.subplots() 
img1 = ax.scatter(x,y, s = sizes , c = colors, cmap  = "viridis", alpha = 0.3 );

fig.colorbar(img1);
```

### ***Bar***
* Vertical
* Horizontal


```python
soft_drink_price = {
    "coke": 10,
    "pepsi": 12,
    "sprite":8
}

fig, ax = plt.subplots()
ax.bar(soft_drink_price.keys() , soft_drink_price.values())
ax.set(title = "Back Hoa Xanh soft drink price" , ylabel = "Price ($)");
```

```python
fig, ax = plt.subplots()
ax.barh(list(soft_drink_price.keys()) , list(soft_drink_price.values()))
ax.set(title = "Back Hoa Xanh soft drink price" , xlabel = "Price ($)");
```

### ***Subplot***

```python
fig , ((ax1,ax2), (ax3,ax4)) = plt.subplots(
    nrows = 2 ,
    ncols = 2 ,
    figsize = (10,5)
)

ax1.plot(x,x/2);
ax1.plot(x,x*2);

ax2.scatter(np.random.random(10),np.random.random(10))

ax3.bar(soft_drink_price.keys(), soft_drink_price.values())

ax4.hist(student_height , bins = 30 );
```

```python
fig , ax = plt.subplots(
    nrows = 2 ,
    ncols = 2 ,
    figsize = (10,5)
)

ax[0][0].plot(x,x/2);   # [][] or [,]
ax[0,0].plot(x,x*2);

ax[0,1].scatter(np.random.random(10),np.random.random(10))

ax[1,0].bar(soft_drink_price.keys(), soft_drink_price.values())

ax[1,1].hist(student_height , bins = 30 );
```

### ***Data Visualization*** 

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
cities = pd.read_csv('california_cities.csv')
cities.head()
```

```python
lat, lon = cities['latd'], cities['longd']
lat[:10] , lon[:10]
```

```python
plt.scatter(lon,lat)
plt.axis('equal')
plt.xlabel('longtitude')
plt.ylabel('latitude');
```

```python
population, area = cities["population_total"], cities["area_total_km2"]

plt.style.use('seaborn-v0_8-whitegrid')
plt.figure(figsize = (8,6))
plt.scatter(
    lon,
    lat,
    c = np.log10(population),
    cmap = 'viridis',
    s = area,
    linewidths=0,
    alpha = 0.5
)
plt.axis('equal')
plt.xlabel('longtitude')
plt.ylabel('latitude');

plt.colorbar(label = 'log$_{10}$(poppulation)')
plt.clim(3,7)

plt.title('Califonia Cities: Population and Area Distribution')
```

```python
cities['area_total_km2'].sort_values(ascending = False)
```

```python
population, area = cities["population_total"], cities["area_total_km2"]

plt.style.use('seaborn-v0_8-whitegrid')
plt.figure(figsize = (8,6))
plt.scatter(
    lon,
    lat,
    c = np.log10(population),
    cmap = 'viridis',
    s = area,
    linewidths=0,
    alpha = 0.5
)
plt.axis('equal')
plt.xlabel('longtitude')
plt.ylabel('latitude');

plt.colorbar(label = 'log$_{10}$(poppulation)')
plt.clim(3,7)

area_range = [50,100,300,500]
for area in area_range:
    plt.scatter([],[],s = area,label = str(area) + 'km$^2$',c = 'k' , alpha = 0.4)
plt.legend(labelspacing = 1 , scatterpoints = 1 , title = 'City Area' )

plt.title('Califonia Cities: Population and Area Distribution')
```

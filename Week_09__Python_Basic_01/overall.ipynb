Em nhận được một dòng log từ hệ thống, được ghi lại dưới dạng một chuỗi duy nhất trông rất lộn xộn. Nhiệm vụ của em là làm sạch và bóc tách nó.

```
dirty_log = "   INFO:2025-10-09;ID=user002;ACTION=login_success;LOCATION=ho chi minh city   "
```

Yêu cầu:

Dùng .strip() để loại bỏ các khoảng trắng thừa ở hai đầu.

Dùng .upper() để chuyển toàn bộ log thành chữ hoa để dễ xử lý.

Kiểm tra xem dòng log này có phải là một thông báo INFO hay không bằng cách dùng .startswith().

Dùng .split(';') để tách chuỗi thành các phần thông tin khác nhau (ví dụ: ['INFO:2025-10-09', 'ID=USER002', ...]).

Từ kết quả ở bước 4, hãy lấy ra thông tin về LOCATION.

Gợi ý: Phần tử này sẽ là 'LOCATION=HO CHI MINH CITY'. Em sẽ cần dùng .split('=') một lần nữa để tách key và value, sau đó lấy value.

Cuối cùng, hãy chuẩn hóa giá trị LOCATION vừa lấy được bằng cách dùng phương thức .title() (một phương thức mới giúp viết hoa chữ cái đầu mỗi từ) và in ra kết quả cuối cùng.

````python
dirty_log = "   INFO:2025-10-09;ID=user002;ACTION=login_success;LOCATION=ho chi minh city   "
dirty_log_modifier = dirty_log.strip().upper()
if dirty_log_modifier.startswith("INFO") : 
    dirty_log_list = dirty_log_modifier.split(";")
    location = None 
    for value in dirty_log_list:
        if value.startswith("LOCATION"):
            location = value.split("=")[1].title()
            break
    if location :
        print(f"Location: {location}") 
    else:
        print("Không tồn tại location")
else:
    print("Object không chứa thông tin info !")
````

- sort tại chỗ (mutate) hoặc dùng sorted(a, key=..., reverse=...).

````python

# ***Excercise:***
# Em đang quản lý điểm thi của một lớp học.
scores = [7.5, 8.0, 5.0, 9.5, 6.5, 8.0]
# Một sinh viên có điểm 5.0 đã xin phúc khảo và được lên 7.0. Em hãy cập nhật điểm này trong danh sách. 
ind = scores.index(5.0)
scores[ind] = 7.0

# Có một sinh viên mới vào lớp với điểm 8.5. Hãy thêm điểm này vào danh sách.
scores.append(8.0)

# Tạo một bản sao của danh sách điểm tên là scores_with_bonus.
# Lớp được cộng 0.5 điểm thưởng cho mỗi sinh viên. 
# Hãy cộng điểm thưởng này vào danh sách scores_with_bonus (dùng vòng lặp for). 
# In cả hai danh sách scores và scores_with_bonus để thấy sự khác biệt.
scores_with_bonus = scores.copy()
scores_with_bonus = [ score + 0.5 if score + 0.5 <= 10 else 10 for score in scores_with_bonus]
print("Bảng điểm gốc:", scores)
print("Bảng điểm sau khi cộng:", scores_with_bonus)

# Sử dụng sorted(), hãy tạo một danh sách mới top_3_scores chứa 3 điểm cao nhất của lớp từ danh sách scores gốc.
top_3_scores = sorted(scores,reverse= True)[:3]
print("Top 3 score: ", top_3_scores)

# Tính điểm trung bình của lớp dựa trên danh sách scores gốc.
score_average = sum(scores)/len(scores)
print("Điểm trung bình dựa trên scores gốc: ", round(score_average,2))
````

````python
def analyze_list(nums):
    if not nums:
        return 0, 0, None
    length_numbers = len(nums)
    sum_numbers = sum(nums)
    max_numbers = max(nums)
    return length_numbers, sum_numbers, max_numbers

# Dữ liệu đầu vào
numbers = [10, 25, 8, 42, 15, 7]

# Gọi hàm và unpacking kết quả vào 3 biến
count, total, maximum = analyze_list(numbers)

# In kết quả
print(f"Số phần tử: {count}")
print(f"Tổng giá trị: {total}")
print(f"Giá trị lớn nhất: {maximum}")
````

````python
# --- Viết hàm của em vào đây ---
def tinh_tong_hoa_don(so_tien_goc, phan_tram_tip=10):
    """Hàm này sẽ nhận vào số tiền của hóa đơn gốc và phần trăm tiền tip (boa),
    sau đó tính và trả về tổng số tiền cuối cùng phải trả."""
    
    tien_tip = so_tien_goc * (phan_tram_tip / 100)
    tong_tien = so_tien_goc + tien_tip
    return tong_tien 

# --- Gọi hàm để kiểm tra ---

# Trường hợp 1: Dùng tip mặc định 10%
hoa_don_1 = tinh_tong_hoa_don(500000)
print(f"Trường hợp 1 - Hóa đơn 500,000 VNĐ, tip mặc định: {hoa_don_1} VNĐ")

# Trường hợp 2: Tip 20%
hoa_don_2 = tinh_tong_hoa_don(500000, 20)
print(f"Trường hợp 2 - Hóa đơn 500,000 VNĐ, tip 20%: {hoa_don_2} VNĐ")
````

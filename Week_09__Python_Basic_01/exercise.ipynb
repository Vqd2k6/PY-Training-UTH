```python
# 1	Nhập tên và năm sinh, in 'Xin chào <tên>, <tuổi> tuổi'.
name = input("Nhập tên: ")
age = int(input("Nhập tuổi: "))
print(f'Xin chào {name}, {age} tuổi')
```

```python
# 2	Tính BMI từ cân nặng(kg) và chiều cao(m), phân loại WHO.
weight = float(input("Nhập cân năng(kg): "))
height = float(input("Nhập chiều cao(m): "))
bmi = weight / (height)**2 

if bmi <= 18.5 :
    print("Thiếu cân")
elif bmi <= 22.9 :
    print("Bình thường")
elif bmi <= 24.9 :
    print("Thừa cân")
elif bmi <= 29.9 :
    print("Gần béo phì")
else:
    print("Béo phì")
```

```python
# 3	Đếm số nguyên tố trong khoảng [a,b].
def is_prime(n):
    if n < 2 :
        return False
    if n <= 3 :
        return True 
    i = 5 
    if n % 2 == 0 or n % 3 == 0 :
        return False 
    while i < n**0.5 :
        if n % i == 0 or n % (i + 2) == 0:
            return False
        i += 6 
    return True 

def count_prime(a , b):
    count = 0 
    for i in range(a,b):
        if is_prime(i):
            count += 1 
    return count 

a = int(input("Số bắt đầu: "))
b = int(input("Số kết thúc: "))
print(f"Số nguyên tố nằm trong khoảng[{a}, {b}] là {count_prime(a,b)}")
```

```python
# 4	Đảo ngược chuỗi nhập vào (không dùng slicing shortcuts ở bài 1).

inp = input("Nhập chuỗi: ")
inp_reverse = "".join([inp[-(i + 1)] for i in range(len(inp))])
print("Chuỗi gốc:", inp )
print("Chuỗi sau khi đảo:", inp_reverse)
```

```python
# 5	Tính trung bình cộng danh sách số, làm tròn 2 chữ số.

numbers = list(map(int,input("Numbers: ").strip().split()))
average = sum(numbers)/len(numbers)

print("Numbers", numbers)
print("Trung bình cộng:", round(average,2))
```

```python
numbers = [2, 4, 1, 0, 3, 9, 5, 8, 7]

max1 = float('-inf')
max2 = float('-inf')

for num in numbers:
    if num > max1:
        max2 = max1
        max1 = num
    elif num > max2 and num != max1: 
        max2 = num

print("Numbers:", numbers)
print("Số lớn thứ 2 trong danh sách là:", max2)

```

```python
# 7	Tách email thành local và domain, kiểm tra định dạng đơn giản.
email = input("Nhập email: ")
if "@" not in email :
    print("Email không đúng định dạng")
else:
    local, domain = email.split("@")
    print("Email:",email)
    print(f"Local: {local}\nDomain: {domain}")


```

```python
def hien_thi_danh_ba(danh_ba):
    if not danh_ba:
        print("Danh bạ rỗng.")
        inp = int(input("Chọn 0 để quay lại\nChọn 1 để kết thúc"))
        return inp 
    else:
        for item in danh_ba:
            print(f"{item["name"]}\t{item["phone"]}")
        inp = int(input("Chọn 0 để quay lại\nChọn 1 để kết thúc"))
        return inp  
    
def them_lien_he(danh_ba):
    name = input("Nhập tên liên hệ: ")
    phone = input("Nhập số điện thoại: ")
    if name and phone:
        danh_ba.append({"name": name, "phone": phone})
        print(f"Đã thêm thành công {name} với sdt {phone} vào danh bạ.")
        inp = int(input("Chọn 0 để quay lại\nChọn 1 để kết thúc"))
        return inp  

    else:
        print("Thông tin nhập không chính xác!")
        inp = int(input("Chọn 0 để quay lại\nChọn 1 để kết thúc"))
        return inp 

def tim_kiem(danh_ba):
    name_find = input("Nhập tên cần tìm: ").strip().lower()
    flag = True 
    for item in danh_ba:
        if name_find == item["name"].lower():
            print(f"{item["name"]}\t{item["phone"]}")
            flag = False 
            break 
    if flag :
        print("Không tìm thấy tên trong danh bạ.")
    inp = int(input("Chọn 0 để quay lại\nChọn 1 để kết thúc"))
    return inp 

def xoa(danh_ba):
    name_del = input("Nhập tên cần xoá: ").strip().lower()
    flag = True 
    ind_del = None
    for index, item in enumerate(danh_ba):
        if name_del == item["name"].lower():
            ind_del = index 
            flag = False 
            break 
    if flag :
        print("Không tìm thấy tên trong danh bạ.")
    else:
        danh_ba.pop(ind_del)
        print(f"Đã xoá thành công {name_del} khỏi danh bạ.")
    inp = int(input("Chọn 0 để quay lại\nChọn 1 để kết thúc"))
    return inp 
```

```python
# 8	Tạo menu quản lý danh bạ (list of dict). Thêm/xóa/tìm theo tên.

def main():
    """Hàm chính điều khiển luồng của chương trình."""
    danh_ba = []

    while True:
        print("\n----- MENU QUẢN LÝ DANH BẠ -----")
        print("1. Hiển thị toàn bộ danh bạ")
        print("2. Thêm liên hệ mới")
        print("3. Tìm kiếm liên hệ theo tên")
        print("4. Xóa liên hệ theo tên")
        print("0. Thoát chương trình")

        lua_chon = input("Vui lòng nhập lựa chọn của bạn: ")

        if lua_chon == '1':
            print("Chức năng hiển thị danh bạ") 
            if not hien_thi_danh_ba(danh_ba) :
                continue
            else:
                break 
        elif lua_chon == '2':
            print("Chức năng thêm liên hệ")
            if not them_lien_he(danh_ba):
                continue
            else:
                break 
        elif lua_chon == '3':
            print("Chức năng tìm kiếm")
            if not tim_kiem(danh_ba) :
                continue
            else:
                break
        elif lua_chon == '4':
            print("Chức năng xóa")
            if not xoa(danh_ba) :
                continue
            else:
                break
        elif lua_chon == '0':
            print("Cảm ơn bạn đã sử dụng chương trình!")
            break # Thoát khỏi vòng lặp while
        else:
            print("Lựa chọn không hợp lệ, vui lòng thử lại.")

main()
```

```python
# 3) Mini Project
# Viết chương trình quản lý chi tiêu: thêm khoản thu/chi, tổng kết theo danh mục, xuất báo cáo dạng bảng văn bản.

# Import thư viện để làm việc với ngày tháng
from datetime import datetime

def them_giao_dich(danh_sach_gd):
    while True :
        try: 
            ngay = datetime.strptime(input("Nhập ngày(YYYY-mm-dd): "), "%Y-%m-%d")
            loai = input("Loại (Thu/Chi): ").strip().lower()
            so_tien = float(input("Số tiền: "))
            danh_muc = input("Danh mục: ").strip()
        except ValueError:
            print("Dữ liệu nhập không hợp lệ !")
        else:
            danh_sach_gd.append({"ngay": ngay,"laoi": loai, "so_tien": so_tien if loai == "thu" else -so_tien, "danh_muc": danh_muc})
            print("Đã thêm giao dịch thành công")
            inp = int(input("Chọn 0 để thoát\nChọn 1 để trở lại\nChọn 2 để tiếp tục thêm"))   #####
            if inp == 2 :
                continue
            else:
                return inp

def tong_ket_theo_danh_muc(danh_sach_gd):
    chi_theo_danh_muc = {}
    thu_theo_danh_muc = {}
    for gd in danh_sach_gd:
        if gd["laoi"].lower() == "thu":
            thu_theo_danh_muc[gd["danh_muc"]] = thu_theo_danh_muc.get(gd["danh_muc"],0) + gd["so_tien"]
        else:
            chi_theo_danh_muc[gd["danh_muc"]] = chi_theo_danh_muc.get(gd["danh_muc"],0) + gd["so_tien"]
    if chi_theo_danh_muc or thu_theo_danh_muc:
        print("Các khoảng thu: ")
        for index,value in thu_theo_danh_muc.items() :
            print(f"Danh mục: {index}\tThu: {value}")
        print("Các khoảng chi: ")
        for index,value in chi_theo_danh_muc.items() :
            print(f"Danh mục: {index}\tChi: {value}")
    else:
        print("Chưa có khoản thu chi nào.")
    inp = int(input("Chọn 0 để thoát\nChọn 1 để trở lại"))    ###
    return inp

def xuat_bao_cao(danh_sach_gd):
    print("=========   BÁO CÁO   =========")
    if danh_sach_gd: 
        tong_thu = 0 
        tong_chi = 0 
        for value in danh_sach_gd:
            print(f"Ngày: {value["ngay"]}\nLoại: {value["laoi"]}\nSố tiền: {value["so_tien"]}\nDanh mục: {value["danh_muc"]}")
            if value["so_tien"] < 0 :
                tong_chi += value["so_tien"]
            else:
                tong_thu += value["so_tien"]
        print("-------------------------------")
        print("Tổng thu:",tong_thu)
        print("Tổng chi:",tong_chi)
        print("Ngân sách còn lại:", tong_chi + tong_thu)
    else:
        print("Chưa có giao dịch nào.")
    inp = int(input("Chọn 0 để thoát\nChọn 1 để trở lại"))     #####
    return inp

def main():
    danh_sach_giao_dich = [] 

    while True:
        print("\n===== CHƯƠNG TRÌNH QUẢN LÝ CHI TIÊU =====")
        print("1. Thêm giao dịch mới (thu/chi)")
        print("2. Thống kê theo danh mục")
        print("3. Xuất báo cáo chi tiết")
        print("0. Thoát")
        print("=" * 42)

        lua_chon = input("Nhập lựa chọn của bạn: ")

        if lua_chon == '1':
            if them_giao_dich(danh_sach_giao_dich) == 0 :
                break 
        elif lua_chon == '2':
            if tong_ket_theo_danh_muc(danh_sach_giao_dich) == 0 :
                break 
        elif lua_chon == '3':
            xuat_bao_cao(danh_sach_giao_dich)
        elif lua_chon == '0':
            print("Tạm biệt!")
            break
        else:
            print("Lựa chọn không hợp lệ. Vui lòng thử lại.")

if __name__ == "__main__":
    main()
```

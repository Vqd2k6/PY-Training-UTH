## ***1) Rẽ nhánh & vòng lặp nâng cao: if/elif/else, for/while, break/continue, else trên vòng lặp***

• if/elif/else: dùng elif thay vì lồng if sâu. Dùng toán tử 3 ngôi rút gọn: a if cond else b.
• for i in range(n): dùng khi biết số bước; while dùng khi lặp theo điều kiện. break dừng vòng lặp; continue bỏ qua phần còn lại của vòng.
• for–else / while–else: khối else chạy khi vòng lặp **không** break (ít người biết nhưng rất hữu ích).
# Ví dụ for–else tìm số nguyên tố
n = 29
for i in range(2, int(n**0.5)+1):
if n % i == 0:
print("Không phải số nguyên tố")
break
else:
print("Là số nguyên tố") # chỉ in khi vòng for không break

```python
# Đây là lô dữ liệu em nhận được
data_batch = [15.2, 16.1, 'N/A', 15.8, 17.3, 200.5, 16.6, 'ERROR', 15.9]

# Ngưỡng an toàn
TEMP_MIN = 10.0
TEMP_MAX = 100.0 # Bất cứ gì trên 100.0 là lỗi cảm biến

for data in data_batch:
    if not isinstance(data, (float,int)):
        continue
    if TEMP_MAX < data or data < TEMP_MIN :
        print(f"Cảnh báo: Phát hiện giá trị ngoài ngưỡng: {data}")
        break
else:
    print("Lô dữ liệu an toàn.")
```

```python
sales_log = [
    {'product_id': 'P001', 'category': 'Electronics', 'quantity': 2},
    {'product_id': 'P002', 'category': 'Clothing', 'quantity': 5},
    {'product_id': 'P003', 'category': 'Books', 'quantity': 10},
    {'product_id': 'P004', 'category': 'Electronics', 'quantity': 1},
    {'product_id': 'P005', 'category': 'Clothing', 'quantity': 3},
    {'product_id': 'P006', 'category': 'Home', 'quantity': 7},
]
category_totals = {}
for data in sales_log:
    key = data["category"]
    category_totals[key] = category_totals.get(key,0) + data["quantity"]
print(category_totals)

category_grouping = {}
for data in sales_log :
    category_grouping.setdefault(data["category"], []).append(data["product_id"])
print(category_grouping)
```

```python
# Danh sách user thấy chiến dịch A
campaign_A_users = ['U001', 'U002', 'U003', 'U004', 'U010']

# Danh sách user thấy chiến dịch B
campaign_B_users = ['U003', 'U004', 'U005', 'U006', 'U007']

# Danh sách user đã đăng ký tài khoản VIP
# Dữ liệu này bị bẩn và có trùng lặp
vip_users_list = ['U001', 'U005', 'U008', 'U001', 'U006']

vip_users = set(vip_users_list)
print(f"Danh sách đăng ký tài khoản vip: {vip_users}")

set_campaign_A_users = set(campaign_A_users)
set_campaign_B_users = set(campaign_B_users)
# Những người thấy cả 2 chiến dịch 
common_users = set_campaign_A_users & set_campaign_B_users 
print(f"Những người thấy cả 2 chiến dịch: {common_users}")
# Những người tiếp xúc ít nhất 1 chiến dịch 
all_reached_users = set_campaign_A_users | set_campaign_B_users
print(f"Những người tiếp xúc ít nhất 1 chiến dịch {all_reached_users}")
# Những người chỉ thấy duy nhất chiến dịch A
a_only_users = set_campaign_A_users - set_campaign_B_users 
print(f"Những người chỉ thấy duy nhất chiến dịch A: {a_only_users}")

# Những user VIP đến từ chiến dịch A (Họ vừa là VIP và vừa thấy camp A)
vip_converts_A = vip_users & set_campaign_A_users
print(f"Những user VIP đến từ chiến dịch A: {vip_converts_A}")

# Những user thấy chiến dịch B nhưng không phải là VIP.
non_vip_B = set_campaign_B_users - vip_users 
print(f"Những user thấy chiến dịch B nhưng không phải là VIP: {non_vip_B}")
```

```python
# Dữ liệu đầu vào
model_names = ["Logistic Regression", "SVM", "Decision Tree"]
model_accuracies = [0.85, 0.92, 0.88]

# Danh sách các feature (đặc trưng) mà mỗi mô hình đã sử dụng
model_features = [
    ["age", "income"],
    ["age", "income", "education", "location"],
    ["age", "location"],
]
# 1) In thông tin mô hình theo định dạng:
# ID 1 - Model: Logistic Regression, Accuracy: 0.85
for index, model in enumerate(
    zip(model_names, model_accuracies), 1
):
    print(f"ID {index} - Model: {model[0]}, Accuracy: {model[1]}")

# 2) Quy tắc 1 (Baseline): Kiểm tra xem "có phải tất cả các mô hình đều có accuracy > 0.80
base_line = all(accuracy > 0.80 for accuracy in model_accuracies)
print(f"Quy tắc 1 (Baseline): {base_line}")

# Quy tắc 2 (Feature Check):
# Kiểm tra xem "có bất kỳ mô hình nào sử dụng feature 'education' không?". In ra True hoặc False
feature_check = any("education" in set(features) for features in model_features)
print(f"Có bất kỳ mô hình nào sử dụng feature 'education' không: {feature_check}")

# 3)
combine_data =  list(zip(model_names,model_accuracies,model_features))
print(
    "Xếp hạng 1 (Theo Accuracy): In ra tên các mô hình, xếp hạng theo accuracy từ cao xuống thấp:"
)


for data in sorted(combine_data, key = lambda data : data[1], reverse = True):
    print(data[0])

print(
    "In ra tên các mô hình, xếp hạng theo số lượng feature (chính là len của list feature) "
    "từ thấp đến cao (ít feature nhất lên trước)."
)
for data in sorted(combine_data, key = lambda data : len(data[2]), reverse = True):
    print(data[0])
```

```python
# Dữ liệu đầu vào
document = """
    This is a great, GREAT course on Python. 
    I learned Python and Data Science. 
    However, the section on 'data visualization' was a bit short.
"""

# Các từ dừng cần loại bỏ
STOP_WORDS = {'a', 'on', 'this', 'is', 'i', 'and', 'the', 'was', 'however'}

import re 
import collections 
tokens = re.findall(r"[\w']+", document.casefold())
filtered_tokens = [token for token in tokens if token not in STOP_WORDS ]

# my_counter = collections.Counter(filtered_tokens)
# my_counter.most_common(3)

my_counter = {}
for token in filtered_tokens:
    my_counter[token] = my_counter.get(token,0) + 1 
print(my_counter)
my_counter_sorted = sorted( my_counter.items(), key = lambda data : data[1] , reverse= True ) 
for i in range(3):
    print(my_counter_sorted[i][0])
```
